```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/Ames_Without_Missing.csv')
```

```python
df
```

```python
df.info()
```

```python
# There are 3 types of categorical features
# 1) Numerical, ordinal
# 2) Numerical, non-ordinal (no linear relationship)
# 3) String, (implied non-ordinal)
# We need to encode categorical features of types 2 and 3
# To separate them out, let's convert type 2 features into strings
```

```python
df['MS SubClass'] = df['MS SubClass'].astype('object')
```

```python
df.info()
```

```python
df_to_encode, df_remaining = df.select_dtypes(include='object'), df.select_dtypes(exclude='object')
```

```python
df_to_encode.shape
```

```python
df_remaining.shape
```

```python
df_to_encode.columns
```

```python
# test_series = pd.Series(['Up', 'Up', 'Down', 'Up', 'Down'])
# pd.get_dummies(test_series, drop_first=True)
```

```python
df_encoded = pd.get_dummies(df_to_encode, drop_first=True)
```

```python
df_encoded
```

```python
df_encoded.shape
```

```python
final_df = pd.concat([df_remaining, df_encoded], axis=1)
```

```python
final_df.shape
```

```python
final_df
```

```python
# Now, we can save this dataframe
```

```python
final_df.to_csv('../DATA/Ames_After_Encoding.csv', index=False)
```

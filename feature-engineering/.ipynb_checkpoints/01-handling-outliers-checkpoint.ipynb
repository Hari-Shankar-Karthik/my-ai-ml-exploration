```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/Ames_Housing_Data.csv')
```

```python
df
```

```python
df.info()
```

```python
df.select_dtypes('number').corr()['SalePrice'].sort_values(key=lambda corr: abs(corr))
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df, x='Overall Qual', y='SalePrice', hue='Gr Liv Area', palette='viridis')
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df, x='Gr Liv Area', y='SalePrice', hue='Overall Qual', palette='viridis')
plt.axhline(y=200000, ls='--', color='black')
```

```python
# We notice some outlier points which have high Overall Qual, high Gr Liv Area but low SalePrice
```

```python
# identifying outliers 
outliers = df[(df['SalePrice'] <= 200000) & (df['Gr Liv Area'] >= 4000)]
outliers
```

```python
# let's check if these outliers are a recent trend
outliers['Yr Sold'].value_counts()
```

```python
df['Yr Sold'].value_counts()
```

```python
# Here, the best course of action would be to drop them so as to improve model performance
```

```python
df = df.drop(index=outliers.index)
```

```python
df
```

```python
df.to_csv('../DATA/Ames_Without_Outliers.csv', index=False)
```

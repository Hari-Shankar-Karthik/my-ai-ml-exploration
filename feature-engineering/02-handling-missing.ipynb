```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/Ames_Without_Outliers.csv')
```

```python
df
```

```python
# Let's drop the PID column (unnecessary index column)
df = df.drop(columns=['PID'])
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0].sort_values()
```

```python
# Only one row is missing the 'Electrical' feature. Moreover, it doesn't have a default value.
# Therefore, best course of action would be to drop that row
df = df.dropna(subset='Electrical')
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
df
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
df.columns
```

```python
# A lot of basement features are missing because some houses don't have a basement
bsmt_NA_cols = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
bsmt_0_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 
               'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_NA_cols] = df[bsmt_NA_cols].fillna(value='NA')
df[bsmt_0_cols] = df[bsmt_0_cols].fillna(value=0)
```

```python
df
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# similarly for garage features
garage_NA_cols = ['Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond']
garage_0_cols = ['Garage Cars', 'Garage Area']
df[garage_NA_cols] = df[garage_NA_cols].fillna(value='NA')
df[garage_0_cols] = df[garage_0_cols].fillna(value=0)
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0].sort_values()
```

```python
# Let's deal with Mas Vnr now, in a similar fashion
df['Mas Vnr Type'] = df['Mas Vnr Type'].fillna('NA')
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)
```

```python
# For Fence, Alley, Misc Feature and Pool QC, we could fill it in with 'NA' values
# Since these are categorical columns, we will have to think about converting them later
# To save us all that hassle, let's just drop them
df = df.drop(columns=['Fence', 'Alley', 'Misc Feature', 'Pool QC'])
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# All the records missing 'Fireplace Qu' don't have a fireplace
df[df['Fireplace Qu'].isna()]['Fireplaces'].value_counts()
```

```python
df['Fireplace Qu'] = df['Fireplace Qu'].fillna(value='NA')
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# Fill in lot frontage based on the description and domain knowledge
# Solution: Fill in missing values with mean per neighbourhood
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value: value.fillna(value.mean()))
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# Some neighbourhoods don't have a lot frontage at all
# Solution: Fill in with 0
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# For garage year built, I'm going to check how many years after the house was it built on average
# Take mean per neighbourhood and fill it in
df['Garage delay'] = df['Garage Yr Blt'] - df['Year Built']
```

```python
df['Garage delay'] = df.groupby('Neighborhood')['Garage delay'].transform(lambda value: value.fillna(value.mean()))
```

```python
df['Garage Yr Blt'] = df['Year Built'] + df['Garage delay']
```

```python
df = df.drop(columns='Garage delay')
```

```python
df
```

```python
percent_missing = 100 * df.isna().sum() / df.shape[0]
percent_missing[percent_missing > 0]
```

```python
# No more missing data! Yay!
# Now save it before it's too late
```

```python
df.to_csv('../DATA/Ames_Without_Missing.csv', index=False)
```

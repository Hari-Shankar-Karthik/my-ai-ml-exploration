```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
X, y = load_digits(return_X_y=True)
```

```python
from sklearn.manifold import SpectralEmbedding
```

```python
X_reduced = SpectralEmbedding(n_components=2).fit_transform(X)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_scaled = MinMaxScaler().fit_transform(X_reduced)
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y, palette='viridis')
plt.title('True clusters')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
for linkage in ['ward', 'average', 'complete', 'single']:
    y_pred = AgglomerativeClustering(n_clusters=10, linkage=linkage).fit_predict(X_scaled)
    fig, ax = plt.subplots()
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='nipy_spectral', ax=ax)
    plt.title(f'{linkage} linkage')
```

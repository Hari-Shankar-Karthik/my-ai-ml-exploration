```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate sample data
n_samples = 1500
np.random.seed(0)
t = 1.5 * np.pi * (1 + 3 * np.random.rand(1, n_samples))
x = t * np.cos(t)
y = t * np.sin(t)

X = np.concatenate((x, y))
X += 0.7 * np.random.randn(2, n_samples)
X = X.T
```

```python
# Scale the data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X_scaled = MinMaxScaler().fit_transform(X)
```

```python
# Connectivity matrix
# Here, we will compute linkages only for the nearest 30 nodes
```

```python
from sklearn.neighbors import kneighbors_graph
```

```python
# The docs use 30 neighbors but that is because the data is unscaled
connectivity_constraints = kneighbors_graph(X_scaled, 10, include_self=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
for n_clusters in [30, 3]:
    for connectivity in [None, connectivity_constraints]:
        fig, axs = plt.subplots(ncols=4, figsize=(12, 4))
        plt.suptitle(f'n_clusters={n_clusters}, connectivity={connectivity is not None}')
        for i, linkage in enumerate(['average', 'complete', 'ward', 'single']):
            clusterer = AgglomerativeClustering(n_clusters=n_clusters, connectivity=connectivity, linkage=linkage)
            y_pred = clusterer.fit_predict(X_scaled)
            # axs[i].scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_pred, cmap=plt.cm.nipy_spectral)
            sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_pred, palette='nipy_spectral', ax=axs[i], lw=0.1)
            axs[i].legend().remove()
            axs[i].set_title(f'linkage={linkage}')
            axs[i].axis('equal')
            axs[i].axis(False)
```

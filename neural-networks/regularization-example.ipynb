```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification, make_moons, make_circles
```

```python
X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=0, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]
```

```python
alpha_range = np.logspace(-1, 1, num=5)
ncols = 1 + alpha_range.shape[0]
fig_height = 5
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
```

```python
fig, ax = plt.subplots(ncols=len(datasets), figsize=(fig_height * len(datasets), fig_height))
plt.suptitle('Given clusterings')
for i in range(len(datasets)):
    X0, y0 = datasets[i]
    sns.scatterplot(x=X0[:, 0], y=X0[:, 1], hue=y0, palette='Set1', ax=ax[i], legend=False)
```

```python
for X0, y0 in datasets:
    fig, ax = plt.subplots(ncols=ncols, figsize=(fig_height * ncols, fig_height))
    sns.scatterplot(x=X0[:, 0], y=X0[:, 1], hue=y0, palette='Set1', ax=ax[0], legend=False)
    for i in range(1, ncols):
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('clf', MLPClassifier(alpha=alpha_range[i - 1], max_iter=2000, random_state=42))
        ])
        pipe.fit(X0, y0)
        x_min, y_min = X0.min(axis=0)
        x_max, y_max = X0.max(axis=0)
        x_axis = np.linspace(x_min, x_max, 50)
        y_axis = np.linspace(y_min, y_max, 50)
        x_coords, y_coords = np.meshgrid(x_axis, y_axis)
        grid = np.vstack((x_coords.reshape(-1), y_coords.reshape(-1))).T
        grid_pred = pipe.predict(grid)
        sns.scatterplot(x=X0[:, 0], y=X0[:, 1], hue=y0, palette='Set1', ax=ax[i], legend=False)
        sns.scatterplot(x=grid[:, 0], y=grid[:, 1], hue=grid_pred, palette='Set1', ax=ax[i], alpha=.25, legend=False)
        ax[i].set_title(f'alpha={alpha_range[i - 1].round(2)}')
```

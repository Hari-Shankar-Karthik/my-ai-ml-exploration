```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True, as_frame=True, scaled=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
```

```python
def compute_rmse(learning_rate, num_weak_learners):
    f0_model = y_train.mean()
    residuals = y_train - f0_model
    gradient_boost_regressor = []
    for m in range(num_weak_learners):
        weak_learner = DecisionTreeRegressor(max_leaf_nodes=20).fit(X_train, residuals)
        gradient_boost_regressor.append(weak_learner)
        residuals -= learning_rate * weak_learner.predict(X_train)
    
    y_pred = f0_model
    for dtr in gradient_boost_regressor:
        y_pred += learning_rate * dtr.predict(X_test)
    
    return root_mean_squared_error(y_test, y_pred)
```

```python
weak_learner_counts = np.arange(20, 31)
rmses = np.vectorize(compute_rmse)(0.1, weak_learner_counts)
sns.lineplot(x=weak_learner_counts, y=rmses)
plt.xlabel('Number of weak learners')
plt.ylabel('Root mean squared error')
```

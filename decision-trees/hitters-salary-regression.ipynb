```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('../datasets/Hitters.csv', index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna(axis='index', subset='Salary')
```

```python
df.info()
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['League', 'Division', 'NewLeague'], drop_first=True)
```

```python
df
```

```python
sns.histplot(df, x='Salary')
```

```python
X, y = df.drop(columns='Salary'), df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
base_model = DecisionTreeRegressor()
```

```python
base_model.get_params()
```

```python
param_grid = {
    'ccp_alpha': np.logspace(-4, 4, num=100)
}
```

```python
grid_model = GridSearchCV(base_model, param_grid, scoring='neg_mean_squared_error', verbose=1)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(X_test)
```

```python
mae, rmse = mean_absolute_error(y_test, y_pred), np.sqrt(mean_squared_error(y_test, y_pred))
```

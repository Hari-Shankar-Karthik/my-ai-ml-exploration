```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
X = dataset.data
y = dataset.target_names[dataset.target]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components=2).fit(X_scaled)
```

```python
X_transformed = pca.transform(X_scaled)
```

```python
sns.scatterplot(x=X_transformed[:, 0], y=X_transformed[:, 1], hue=y, palette='Set1')
plt.xlabel(f'PC1 ({(pca.explained_variance_ratio_[0] * 100).round(1)}% var)')
plt.ylabel(f'PC2 ({(pca.explained_variance_ratio_[1] * 100).round(1)}% var)')
for i in range(pca.components_.shape[1]):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], width=.01, head_width=.1, head_length=.1, color='black')
    plt.text(pca.components_[0, i] * 1.1, pca.components_[1, i] * 1.1, dataset['feature_names'][i], fontdict={ 'size': 8 })
```

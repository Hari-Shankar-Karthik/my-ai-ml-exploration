```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
dataset = load_digits()
```

```python
# print(dataset['DESCR'])
```

```python
X = dataset['data']
y = dataset['target_names'][dataset['target']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2).fit(X_scaled)
```

```python
X_transformed = pca.transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
sns.scatterplot(x=X_transformed[:, 0], y=X_transformed[:, 1], hue=y.astype(str), palette='nipy_spectral')
plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel(f'PC1 ({(explained_variance[0] * 100).round(1)}% var)')
plt.ylabel(f'PC2 ({(explained_variance[1] * 100).round(1)}% var)')
```

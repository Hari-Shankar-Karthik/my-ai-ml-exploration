```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../datasets/Advertising.csv')
```

```python
df
```

```python
X, y = df.drop(columns='Sales'), df['Sales']
```

```python
poly_conv = PolynomialFeatures(degree=3, include_bias=False)
```

```python
X = poly_conv.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)
```

```python
base_model = ElasticNet()
```

```python
base_model.get_params()
```

```python
hyperparam_grid = {
    'alpha': [0.01, 0.1, 1, 10, 100],
    'l1_ratio': [0.01, 0.1, 0.5, 0.9, 0.99]
}
```

```python
grid_model = GridSearchCV(base_model, hyperparam_grid, scoring='neg_root_mean_squared_error', cv=10)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_predicted = grid_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_predicted))
```

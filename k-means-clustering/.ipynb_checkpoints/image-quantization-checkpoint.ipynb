```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_sample_image
```

```python
image = load_sample_image('china.jpg')
```

```python
def plot_image(image, title=None, ax=None):
    if ax is not None:
        ax.imshow(image)
        ax.axis('off')
        if title is not None:
            ax.set_title(title)
    else:
        plt.imshow(image)
        plt.axis('off')
        if title is not None:
            plt.title(title)
```

```python
plot_image(image, title='Original image')
```

```python
from sklearn.cluster import KMeans
```

```python
def quantize_image(image, n_colors=64):
    original_pixels = image.reshape(-1, 3)
    clusterer = KMeans(n_clusters=n_colors)
    labels = clusterer.fit_predict(original_pixels)
    palette = clusterer.cluster_centers_.round(0).astype('int')
    return palette[labels].reshape(image.shape)
```

```python
from sklearn.utils import shuffle
from sklearn.metrics import pairwise_distances_argmin
```

```python
def randomly_quantize_image(image, n_colors=64, random_state=None):
    original_pixels = image.reshape(-1, 3)
    palette = shuffle(original_pixels, n_samples=n_colors, random_state=random_state)
    labels = pairwise_distances_argmin(original_pixels, palette)
    return palette[labels].reshape(image.shape)
```

```python
n_colors = 64
quantized_image = quantize_image(image, n_colors=n_colors)
randomly_quantized_image = randomly_quantize_image(image, n_colors=n_colors, random_state=42)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 8), dpi=200)
plot_image(image, ax=ax1, title=f'Original image ({np.unique(image.reshape(-1, 3), axis=0).shape[0]} colors)')
plot_image(quantized_image, ax=ax2, title=f'Quantized image ({n_colors} colors, K-Means)')
plot_image(randomly_quantized_image, ax=ax3, title=f'Quantized image ({n_colors} colors, Random)')
plt.tight_layout()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/Hitters.csv', index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna(subset='Salary')
```

```python
df.info()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['League', 'Division', 'NewLeague'], drop_first=True)
```

```python
df.head()
```

```python
# The data expert has informed us that it might be easier to predict log(salary) instead of salary
```

```python
df['Log Salary'] = np.log(df['Salary'])
```

```python
df.head()
```

```python
X = df.drop(columns=['Salary', 'Log Salary'])
y = df['Log Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, root_mean_squared_error
```

```python
def score_func(y_true, y_pred):
    return root_mean_squared_error(np.exp(y_true), np.exp(y_pred))

custom_scorer = make_scorer(score_func, greater_is_better=False)
```

```python
grid = GridSearchCV(RandomForestRegressor(), {
    'n_estimators': [100, 200, 500],
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
    'max_features': ['sqrt', 'log2']
}, scoring=custom_scorer, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
score_func(y_test, y_pred)
```

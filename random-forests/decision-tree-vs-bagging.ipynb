```python
# AIM: To create a BaggingClassifier from a DecisionTreeClassifier and compare performance
# [Similarly for regression too] 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
# Get the data, clean it and stuff
```

```python
df = pd.read_csv('../datasets/Heart.csv', index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['Sex', 'ChestPain', 'Fbs', 'RestECG', 'ExAng', 'Slope', 'Ca', 'Thal'])
```

```python
df
```

```python
X, y = df.drop(columns='AHD'), df['AHD']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# Create the best Decision Tree you possibly can
```

```python
dt = DecisionTreeClassifier()
```

```python
dt_grid = GridSearchCV(dt, { 'ccp_alpha': np.logspace(-4, 4, num=100) }, scoring='accuracy', verbose=3)
```

```python
dt_grid.fit(X_train, y_train)
```

```python
dt_grid.best_params_
```

```python
dt_pred = dt_grid.predict(X_test)
```

```python
accuracy_score(y_test, dt_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(dt_grid, X_test, y_test)
```

```python
# Now, let's use BaggingClassifier
```

```python
bag = BaggingClassifier(estimator=dt)
```

```python
bag_grid = GridSearchCV(bag, { 'n_estimators': np.arange(100, 1001, 100) }, scoring='accuracy', verbose=3)
```

```python
bag_grid.fit(X_train, y_train)
```

```python
bag_grid.best_params_
```

```python
bag_pred = bag_grid.predict(X_test)
```

```python
accuracy_score(y_test, bag_pred)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/Heart.csv', index_col=0)
```

```python
df
```

```python
# AIM: To make a RandomForestClassifier to predict AHD, and compare the importances of the features
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.head()
```

```python
categorical_cols = ['Sex', 'ChestPain', 'Fbs', 'RestECG', 'ExAng', 'Slope', 'Ca', 'Thal']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
df.head()
```

```python
X, y = df.drop(columns='AHD'), df['AHD']
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# NOTE: Modified from the original
rfc = RandomForestClassifier(criterion='log_loss', max_features='log2')
param_grid = {
    'n_estimators': np.arange(400, 600, 10),
}
```

```python
grid = GridSearchCV(rfc, param_grid, scoring='accuracy', verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(grid, X_test, y_test)
```

```python
RocCurveDisplay.from_estimator(grid, X_test, y_test)
```

```python
plt.figure(dpi=200)
feature_importances = pd.DataFrame({
    'Feature': grid.feature_names_in_, 
    'Importance': grid.best_estimator_.feature_importances_
}).sort_values(by='Importance')
sns.barplot(feature_importances, x='Importance', y='Feature')
```

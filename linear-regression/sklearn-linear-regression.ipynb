```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
```

```python
df = pd.read_csv('../datasets/Advertising.csv')
```

```python
df
```

```python
# feature matrix and label vector
X, y = df.drop(columns='Sales'), df['Sales']
```

```python
X
```

```python
y
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
# Train the model using linear regression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# test the model
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mae
```

```python
rmse
```

```python
y_test.mean()
```

```python
rmse / y_test.mean()
```

```python
# now, let's plot out the residuals and check for discrepancies
residuals = y_test - predictions
```

```python
sns.kdeplot(residuals)
residuals.median()
```

```python
sns.scatterplot(x=y_test, y=residuals)
plt.axhline(y=0, color='black', ls='--')
```

```python
# This plot shows us that the model works best between 10 and 20 units
```

```python
sns.residplot(x=y_test, y=predictions) # same plot, seaborn-style
# NOTE: here, +ve means that the model overshot, -ve means that the model undershot
```

```python
# let's visualize how the residuals fare w.r.t. the normal distribution
fig, ax = plt.subplots()
_ = sp.stats.probplot(residuals, plot=ax)
```

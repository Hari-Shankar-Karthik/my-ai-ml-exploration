```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/Advertising.csv')
```

```python
df
```

```python
# I want to regress Sales onto TV Advertisement and also specify the confidence intervals
```

```python
X = df['TV']
y = df['Sales']
n = X.shape[0]
```

```python
temp = np.sum((X - X.mean()) ** 2)
```

```python
# Linear regression coefficient estimates
beta_1_hat = np.sum((X - X.mean()) * (y - y.mean())) / temp
beta_0_hat = y.mean() - beta_1_hat * X.mean()
```

```python
# Estimating residual standard error
rse_squared = np.sum((y - (beta_0_hat + beta_1_hat * X)) ** 2) / (n - 2)
```

```python
# Estimating standard errors in each of the coefficients
beta_0_se = np.sqrt(rse_squared * ((1/n) + (X.mean() ** 2 / temp)))
beta_1_se = np.sqrt(rse_squared / temp)
```

```python
# Plotting the regression line and confidence intervals
# Confidence interval: 95% chance that the true line exists in this interval
# The confidence interval is two standard errors on either side of the calculated value
sns.scatterplot(x=X, y=y)
sns.lineplot(x=X, y=beta_0_hat + beta_1_hat * X, color='blue')
sns.lineplot(x=X, y=beta_0_hat + 2 * beta_0_se + (beta_1_hat - 2 * beta_1_se) * X, color='red')
sns.lineplot(x=X, y=beta_0_hat - 2 * beta_0_se + (beta_1_hat + 2 * beta_1_se) * X, color='red')
```

```python
# For comparison
sns.regplot(x=X, y=y)
```

```python
(beta_0_hat - 2 * beta_0_se, beta_0_hat + 2 * beta_0_se)
```

```python
(beta_1_hat - 2 * beta_1_se, beta_1_hat + 2 * beta_1_se)
```

```python
# Therefore, in the absence of TV advertisements, the company will sell between 6,117 and 7,948 units
# For every additional $1,000 spent on TV advertisements, the company will sell between 42 and 53 units
# Both the above statements are said with 95% certainty.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/Advertising.csv')
```

```python
df
```

```python
# obtain the feature matrix
X = df.drop(columns='Sales').to_numpy()
X = np.hstack((np.ones((X.shape[0], 1)), X))
```

```python
# obtain the label vector
y = df['Sales'].to_numpy()
```

```python
# compute the coefficient matrix
beta = np.linalg.multi_dot(((np.linalg.inv(np.matmul(X.T, X))), X.T, y))
```

```python
beta
```

```python
df['Predicted Sales'] = np.matmul(X, beta)
```

```python
df
```

```python
df['Residuals'] = df['Predicted Sales'] - df['Sales']
```

```python
sns.histplot(data=df['Residuals'], kde=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/Advertising.csv')
```

```python
df
```

```python
# regress sales onto advertising budget
```

```python
x = df['TV'].to_numpy()
```

```python
y = df['Sales'].to_numpy()
```

```python
beta_1 = np.sum((x - x.mean()) * (y - y.mean())) / np.sum((x - x.mean()) ** 2)
```

```python
beta_0 = y.mean() - beta_1 * x.mean()
```

```python
sns.scatterplot(data=df, x='TV', y='Sales')
plt.plot(df['TV'], beta_0 + beta_1 * df['TV'])
```

```python
def plot_univariate_linear_regression(df, feature_col, label_col, title=''):
    x, y = df[feature_col].to_numpy(), df[label_col].to_numpy()
    beta_1 = np.sum((x - x.mean()) * (y - y.mean())) / np.sum((x - x.mean()) ** 2)
    beta_0 = y.mean() - beta_1 * x.mean()
    sns.scatterplot(df, x=feature_col, y=label_col)
    plt.plot(df[feature_col], beta_0 + beta_1 * df[feature_col])
    plt.title(title)
```

```python
plot_univariate_linear_regression(df, 'TV', 'Sales', 'Variation of Sales with TV Advertisements')
```

```python
plot_univariate_linear_regression(df, 'Radio', 'Sales', 'Variation of Sales with Radio Advertisements')
```

```python
plot_univariate_linear_regression(df, 'Newspaper', 'Sales', 'Variation of Sales with Newspaper Advertisements')
```

```python
# corrected implementation of the DBSCAN algorithm
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels_true, cmap=plt.cm.nipy_spectral)
plt.title('Ground truth clusters')
```

```python
from collections import deque
```

```python
def dbscan(X, eps=.5, min_samples=5):
    # -3 -> unvisited point
    # -2 -> visited, unlabelled point
    # -1 -> outlier
    labels_pred = np.full(X.shape[0], -3)
    
    # for marking clusters
    cluster_id = 0
    
    # compute distances from each point to every other point
    # note that there exists an sklearn function which does this for you
    N, D = X.shape
    X1 = X.reshape(N, 1, D)
    X2 = X.reshape(1, N, D)
    pairwise_distances = np.sqrt(((X1 - X2) ** 2).sum(axis=-1))
    
    queue = deque()
    
    # looping while there are still unvisited datapoints
    while np.any(labels_pred == -3):
        # randomly select an unvisited point to explore its cluster
        cluster_init = np.random.choice(np.where(labels_pred == -3)[0])
        cluster_size = 0
        queue.append(cluster_init)
        
        while queue:
            current = queue.popleft()
            labels_pred[current] = cluster_id
            cluster_size += 1
    
            # check the neighbors
            neighbors = np.where(pairwise_distances[current] <= eps)[0]

            if neighbors.shape[0] >= min_samples:
                # current point is core, extend the cluster to all its neighbors
                # even if a neighbor was previously labelled -1, consider it here
                # [prior exploration of border points labels them as outliers]
                new_neighbors = neighbors[np.isin(labels_pred[neighbors], [-1, -3])]
                labels_pred[new_neighbors] = -2
                queue.extend(new_neighbors)
    
        if cluster_size == 1:
            labels_pred[cluster_init] = -1
        else:
            cluster_id += 1

    return labels_pred
```

```python
labels_pred = dbscan(X, eps=.4, min_samples=20)
plt.scatter(X[:, 0], X[:, 1], c=labels_pred)
plt.title('Custom DBSCAN function')
print(f'n_clusters = {np.unique(labels_pred[labels_pred >= 0]).shape[0]}')
print(f'n_outliers = {(labels_pred == -1).sum()}')
```

```python
from sklearn.cluster import DBSCAN
```

```python
labels_pred = DBSCAN(eps=.4, min_samples=20).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels_pred)
plt.title('sklearn.cluster.DBCSAN')
print(f'n_clusters = {np.unique(labels_pred[labels_pred >= 0]).shape[0]}')
print(f'n_outliers = {(labels_pred == -1).sum()}')
```

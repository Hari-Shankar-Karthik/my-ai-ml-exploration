```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
# Define centers for the blobs
centers = [[2, 2], [4, 5], [10, 10]]  # First two blobs are close to each other

# Generate the blobs
X, y = make_blobs(n_samples=300, centers=centers, cluster_std=1.0, random_state=42)
```

```python
def scatterplot(ax, X, y, title):
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='black', lw=.5, cmap=plt.cm.nipy_spectral)
    ax.set_title(title)
```

```python
fig, ax = plt.subplots()
scatterplot(ax, X, y, 'Ground-truth clusters')
```

```python
# KMeans vs. DBSCAN
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
kmeans_pred = KMeans(n_clusters=3, n_init=10).fit_predict(X)
dbscan_pred = DBSCAN(eps=.475, min_samples=5).fit_predict(X)
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))
scatterplot(ax1, X, kmeans_pred, 'KMeans')
scatterplot(ax2, X, dbscan_pred, 'DBSCAN')
```
